```python
## сначала нам надо загрузить данные 
import pandas as pd #
titanic = pd.read_csv ('titanic_age_class_sex.csv', sep=';')
titanic
```

```python
numeric_sex = pd.get_dummies(titanic['sex']) ## get_dummies позволяет превратить нечисловые данные в числовые
numeric_sex
```

```python
titanic['sex'] = numeric_sex['female'] ## заменим sex на нолики для мальчиков и единички для девочек
titanic
```

```python
target = titanic['survived'] ## столбец выжил/умер -- наша целевая переменная
```

```python
data = titanic[['sex','age','pclass']] ## наши данные (признаки), на которых будем учить классификатор
```

```python
## небольшое техническое преобразование (из DataSeries в numpy-array), не обращайте внимания
data_for_classifier = data.as_matrix() 
target_for_classifier = target.as_matrix()
```

```python
## теперь можно обучать классификатор
from sklearn.tree import DecisionTreeClassifier## импортируем классификатор на деревьях решений
moy_klassifikator = DecisionTreeClassifier()# создаем конкретный объект класса DecisionTreeClassifier()
moy_klassifikator.fit (data_for_classifier, target_for_classifier) # обучаем. сейчас мы хотим 
#не тестировать классификатор, а получить дерево ,так что делаем его сразу на всех данных, без разделения
```

```python
from sklearn.tree import export_graphviz
from graphviz import Source
from IPython.display import SVG
graph = Source(export_graphviz(moy_klassifikator, out_file=None, class_names =['died','survived'],feature_names=['sex','age','pclass']))
SVG(graph.pipe(format='svg'))
```

```python
from sklearn.model_selection import train_test_split ## импортируем функцию для разделения на тренировочную и тестовую выборки
titanic_data_train, titanic_data_test, target_train, target_test = train_test_split(data_for_classifier, target_for_classifier) # получаем тренировочные и тестовые данные
print (len(titanic_data_train)) ## сколько пассажиров в тренировочной
print (len(titanic_data_test)) ## сколько пассажиров в тестовой
```

```python
print ('Качество')
moy_klassifikator.score(titanic_data_test,target_test)
```
